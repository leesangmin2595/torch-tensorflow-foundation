## 함수 정의

```python
def fizzbuzz(max_num):
    counter=torch.tensor(0)
    max_num= torch.tensor(max_num)
    for num in range(1,max_num.numpy()+1):
        num = torch.tensor(num)
        if int(num % 3) == 0 and int(num % 5) == 0:
            print('3과 5의 배수')
        elif int(num % 3) == 0:
            print('3의 배수')
        elif int(num % 5) == 0:
            print('5의 배수')
        else:
            print(num.numpy())
        counter += 1
```

## 함수 실행

```python
fizzbuzz(15)
```

## gradient 계산

```python
print('Computing gradients')
W=torch.tensor([1.0],requires_grad=True)
Y=W*W
Y.backward()
print(W.grad)
```
