```python
import torch
import numpy as np
```

Tensor Constrution

1. Directly from data

```python
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)
```

```python
print(type(data), type(x_data))
print(x_data)
```

2. From a NumPy array

```python
np_array = np.array(data)
x_torch = torch.from_numpy(np_array)
```

```python
print(type(data), type(np_array), type(x_torch))
print(x_data)
```

3. From another tensor

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

4. With random or constant values

```python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

Tensor Attributes

  - dim (rank), size (shape), dtype, device

Tensor : dim & size

```python
tensor = torch.tensor([[1, 2],[3, 4]])
print(tensor)

print(f"Dim of tensor: {tensor.dim()}. or {tensor.ndim}")
print(f"Sizee(Shape) of tensor: {tensor.shape}, or {tensor.size()}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
tensor = torch.rand(3, 4, 5)

print(tensor)
print(f"Dim of tensor: {tensor.dim()}, or {tensor.ndim}")
print(f"Sizee(Shape) of tensor: {tensor.shape}, or {tensor.size()}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

Tensor : Datatype

```python
tensor = torch.FloatTensor([[1, 2],[3, 4]])

print(tensor)
print(f"Dim of tensor: {tensor.dim()}")
print(f"Sizee(Shape) of tensor: {tensor.shape}, or {tensor.size()}")
print(f"Datatype of tensor: {tensor.dtype}")
```

```python
tensor = torch.IntTensor([[1, 2],[3, 4]])

print(tensor)
print(f"Datatype of tensor: {tensor.dtype}")
```

```python
tensor = torch.IntTensor([[1, 2],[3, 4]]).to(torch.float32)

print(tensor)
print(f"Datatype of tensor: {tensor.dtype}")
```

Datatype Promotion

```python
float_tensor = torch.ones(1, dtype=torch.float)
double_tensor = torch.ones(1, dtype=torch.double)
int_tensor = torch.ones(1, dtype=torch.int)
long_tensor = torch.ones(1, dtype=torch.long)
bool_tensor = torch.ones(1, dtype=torch.bool)

print(float_tensor.dtype, double_tensor.dtype, int_tensor.dtype, long_tensor.dtype, bool_tensor.dtype)
```

```python
print(torch.add(5, 5).dtype)
print((int_tensor + 5).dtype)

print((long_tensor + int_tensor).dtype)
print((bool_tensor + long_tensor).dtype)
print((bool_tensor + int_tensor).dtype)

print((float_tensor + double_tensor).dtype)
print(torch.add(long_tensor, float_tensor).dtype)
```

Tensor : Device

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using "{}" device'.format(device))
```

```python
x = torch.randn(2, 3)
print(x)

x = x.to(device) # change device using to() method
y = torch.ones_like(x, device=device)   # directly create y
print(y)
z = x + y
print(z)
print(z.to("cpu", torch.double))
```

```python
device_0 =torch.device('cuda')  # current cuda device
device_1 = torch.device('cuda:0')
device_2 = torch.device('cuda', 0)

print(device_0, device_1, device_2)

device_3 = torch.device('cpu')
device_4 = torch.device('cpu', 0)

print(device_3, device_4)

cuda_0 = torch.device('cuda:0')
x = torch.randn((2,3), device=cuda_0)
print(x)

# You can substitute the torch.device with a string
y = torch.randn((2,3), device='cuda:0')
print(y)
```

Tensor vs NumPy

Tensor to NumPy array

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")

print(type(t), type(n))
```

```python
t.add_(1) # A change in the tensor reflects in the NumPy array.
print(f"t: {t}")
print(f"n: {n}")
print(type(t), type(n))
```

NumPy array to Tensor

```python
n = np.ones(5)
t = torch.from_numpy(n)

np.add(n, 1, out=n) # Changes in the NumPy array reflects in the tensor.
print(f"t: {t}")
print(f"n: {n}")
print(type(t), type(n))
```

Tensor Gradients *** Use requires_grad=True***

image.png

```python
x = torch.tensor([[1., 0.], [-1., 1.]], requires_grad=True)  # 1
print(x)

z = x.pow(2).sum() # 2
z.backward() # 3
print(x.grad) # 4
```
