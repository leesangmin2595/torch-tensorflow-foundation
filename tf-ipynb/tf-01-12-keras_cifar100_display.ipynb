```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Load CIFAR-100 data

```python
(X_train, y_train),(X_test,y_test) = tf.keras.datasets.cifar100.load_data()
print("Training Data :","X_train",X_train.shape,"y_train", y_train.shape, "\n",
      "Test Data :","X_test", X_test.shape,"y_test", y_test.shape)
```

### Print CIFAR-10 data (3-Channel)

```python
# train_data 0
for k in range(3):
    print("Channel : %d"%k)
    for i in range(32):
        for j in range(32):
            print('%3d'%X_train[0,i,j,k], end=' ')
    print()
```

### CIFAR-100 data & target Display

```python
fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True, sharey=True,)
ax = ax.flatten()
for i in range(25):
    img = X_train[i]
    ax[i].imshow(img, interpolation='nearest')
#    ax[i].imshow(img, interpolation='nearest')
    ax[i].set_title(' t: %d' % (y_train[i]))

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```
