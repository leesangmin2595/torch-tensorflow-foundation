## Multivariable Linear Regression : Keras

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Step 1. Data

```python
x_data = np.array([[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]])
y_data = np.array([[152.],
          [185.],
          [180.],
          [196.],
          [142.]])
```

### Step 2. Linear Regression Model (input = 3, output = 1)

```python
epoch=1000
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_dim=3))
```

### Step 3. Loss,Optimizer, Metric

```python
model.compile(loss='mse', optimizer='rmsprop')
```

### Step 4. Train the model

```python
history = model.fit(x_data, y_data, epochs=epoch)
```

### Step 5. Test the model

```python
y_predict = model.predict(np.array([[95., 100., 80]]))
print(y_predict)
```

### Plot the Training loss

```python
epochs = np.arange(1, epoch+1)
plt.plot(epochs, history.history['loss'], label='Training loss')
#plt.plot(epochs, history.history['val_loss'], label='Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
