## TF2에서 TF1 코드를 실행하기 위한 명령문

```python
import tensorflow.compat.v1 as tf
```

```python
tf.disable_v2_behavior()
```

## TF1에서 계산그래프 구성

```python
W = tf.Variable([2.])
b = tf.Variable([1.])
x = tf.placeholder(tf.float32)
```

```python
h = tf.nn.relu(W*x + b)
```

## TF1에서 session 및 변수 초기화

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

## TF1에서는 실행을 위해 sess.run() 함수 사용

```python
print("W = ", sess.run(W))
print("b = ", sess.run(b))
print("h = ", sess.run(h, feed_dict={x:[1., -1.]}))
```
