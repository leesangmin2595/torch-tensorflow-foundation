## W, b, x 정의

```python
W = tf.Variable([2.])
b = tf.Variable([1.])
x = tf.constant([1., -1.])
```

```python
print("W = ", W.numpy())
print("b = ", b.numpy())
```

## h = Wx + b 계산

```python
h = tf.nn.relu(W*x + b)
```

```python
print('x = ', x.numpy(), 'h = ', h.numpy())
```
